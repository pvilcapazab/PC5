```python
import pandas as pd

df = pd.read_csv("./data/winemag-data-130k-v2.csv")
df.rename(columns={'country': 'pais', 'description': 'descripcion', 'points': 'puntos', 'price': 'precio'}, inplace=True)
df.head()
```

```python
continentes = {
    'Europa': ['Italy', 'Portugal', 'Spain', 'France', 'Germany', 'Austria',
               'Hungary', 'Greece', 'Romania', 'Czech Republic', 'Slovenia',
               'Luxembourg', 'Croatia', 'Serbia', 'Moldova', 'England', 'Switzerland',
               'Bosnia and Herzegovina', 'Ukraine', 'Slovakia', 'Macedonia'],
    'Norteamérica': ['US', 'Mexico', 'Canada'],
    'Sudamérica': ['Argentina', 'Chile', 'Uruguay', 'Brazil', 'Peru'],
    'Oceanía': ['Australia', 'New Zealand'],
    'Asia': ['Israel', 'Turkey', 'Georgia', 'Lebanon', 'India', 'Armenia'],
    'África': ['South Africa', 'Morocco']
}

def etiquetar_continente(pais):
    for continente, paises in continentes.items():
        if pais in paises:
            return continente
    return 'Otro'

def etiquetar_precio(precio:float):
    if precio < 20:
        return 'Precio accesible'
    elif 20 <= precio < 50:
        return 'Precio moderado'
    elif 50 <= precio < 100:
        return 'Precio moderadamente alto'
    elif 100 <= precio <= 500:
        return 'Precio muy caro'
    elif precio > 500:
        return 'Precio demasiado caro'
    else:
        return 'Precio desconocido'
    
def etiquetar_estrellas(puntos: int):
    if puntos == 80:
        return '1 estrella'
    elif 80 < puntos <= 85:
        return '2 estrellas'
    elif 85 < puntos < 90:
        return '3 estrellas'
    elif 90 <= puntos <= 99:
        return '4 estrellas'
    elif puntos == 100:
        return '5 estrellas'

df['continentes'] = df.pais.apply(etiquetar_continente)
df['etiqueta_costo'] = df.precio.apply(etiquetar_precio)
df['etiqueta_estrellas'] = df.puntos.apply(etiquetar_estrellas)

df
```

Genere 4 reportes por agrupamiento de datos. Deberá elegir que reportes realizar
Ejemplo1: Por contienente mostrar los vinos mejor puntuados
Ejemplo2: Promedio de precio de vino y cantidad de reviews obtenidos según pais. Ordenado de mejor a peor

```python
reporte_1 = df.groupby('continentes').agg({'puntos': 'mean'}).sort_values(by='puntos', ascending=False).reset_index()
reporte_1.columns = ['Continente', 'Promedio de Puntos de Vinos']
print(reporte_1)

reporte_2 = df.groupby('pais').agg({'descripcion': 'count', 'precio': 'mean'}).sort_values(by='descripcion', ascending=False).reset_index()
reporte_2.columns = ['País', 'Cantidad de Reviews', 'Promedio de Precios de Vinos']
print(reporte_2)
```

```python
reporte_3 = df.groupby(['pais', 'variety']).size().reset_index(name='count')
reporte_3 = reporte_3.sort_values(by=['pais', 'count'], ascending=False).groupby('pais').agg({'variety': 'first', 'count': 'first'}).reset_index()
reporte_3.columns = ['País','Vino más común', 'Cantidad']
print(reporte_3)
```

```python
reporte_4 = df.groupby(['continentes', 'variety']).size().reset_index(name='count')
reporte_4 = reporte_4.sort_values(by=['continentes', 'count'], ascending=False).groupby('continentes').agg({'variety': 'first', 'count': 'first'}).reset_index()
reporte_4.columns = ['Continente','Vino más común', 'Cantidad']
print(reporte_4)
```
